## Inaficent Way to sort letters
- Has to loop through all letters
- Then Compare
- Then swap

```python
letters = [["a", "b", "c"], ["g", "h", "i"], ["x", "y", "z"]]

def bubble_sort(letters: List[str]) -> List[str]:
    # Loop through the list of letters
    for i in range(len(letters)):
        # Compare adjacent elements
        for j in range(i + 1, len(letters)):
            # If the elements are out of order, swap them
            if letters[i] > letters[j]:
                letters[i], letters[j] = letters[j], letters[i]
    # Return the sorted list
    return letters

sorted_letters = bubble_sort(letters)
print(sorted_letters)
```

## More efficent way
- Only goes to 2 items
- Then Compare
- Then swap
- Will be faster on larger data pools

```python
letters = [["a", "b", "c"], ["g", "h", "i"], ["x", "y", "z"]]

def quicksort(letters: List[List[str]]) -> List[List[str]]:
    # If the list has fewer than 2 elements, it is already sorted
    if len(letters) < 2:
        return letters
    # Select a pivot element
    pivot = letters[0]
    # Partition the list into smaller and larger groups
    smaller = [x for x in letters if x[0] < pivot[0]]
    larger = [x for x in letters if x[0] > pivot[0]]
    # Sort the smaller and larger groups recursively
    smaller_sorted = quicksort(smaller)
    larger_sorted = quicksort(larger)
    # Concatenate the smaller group, the pivot, and the larger group
    return smaller_sorted + [pivot] + larger_sorted

sorted_letters = quicksort(letters)
print(sorted_letters)
```
